```python
class learn():
    
    pred_col = '_predicted'
    
    def __init__(self,features,target,cv):
        
        self.features = features
        self.target = target
        self.series_index = target.index
        self.rename_col = target.columns[0]
        
        df = features.join(target)

        self.cv = cv
        self.pred_col
        
        from sklearn.model_selection import train_test_split
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.features, self.target, test_size=0.2, random_state=41234)

        self.series_index_split = self.y_test.index

        df_norm = normalize(df)
        self.features_norm = df_norm[features.columns]
        self.target_norm = df_norm[target.columns]
        self.X_train_norm, self.X_test_norm, self.y_train_norm, self.y_test_norm = train_test_split(self.features_norm, self.target_norm, test_size=0.2, random_state=41234)
        
        self.series_index_norm = self.target_norm.index
        self.series_index_split_norm = self.y_test_norm.index

        
    
    def linear(self,cross_validate=True):
        
        from sklearn import linear_model
        reg = linear_model.LinearRegression()
        

        if cross_validate == True:
            predictions_frame = cross_val_regression(reg,self.features,self.target,self.series_index,self.rename_col,self.pred_col,self.cv)
            
        else:
            predictions_frame = train_test_regression(reg,self.X_train,self.X_test,self.y_train,self.y_test,self.series_index_split,self.rename_col,self.pred_col)
            
        coef_v, mbe, r2 = get_metrics(predictions_frame)
            
        return predictions_frame, coef_v , mbe, r2
    
    def rdf(self,cross_validate=True):
        
        from sklearn.ensemble import RandomForestRegressor
        reg = RandomForestRegressor()
        
        if cross_validate == True:
            predictions_frame = cross_val_regression(reg,self.features,self.target,self.series_index,self.rename_col,self.pred_col,self.cv)
            
        else:
            predictions_frame = train_test_regression(reg,self.X_train,self.X_test,self.y_train,self.y_test,self.series_index_split,self.rename_col,self.pred_col)
       
        coef_v, mbe, r2 = get_metrics(predictions_frame)
            
        return predictions_frame, coef_v , mbe, r2
    
    def svm(self,cross_validate=True):
        
        from sklearn.svm import SVR
        reg = SVR(kernel='rbf')
        
        if cross_validate == True:
            predictions_frame = cross_val_regression(reg,self.features_norm,self.target_norm,self.series_index_norm,self.rename_col,self.pred_col,self.cv)
            
        else:
            predictions_frame = train_test_regression(reg,self.X_train_norm,self.X_test_norm,self.y_train_norm,self.y_test_norm,self.series_index_split_norm,self.rename_col,self.pred_col)
            
        #denormalize the dataframe
        predictions_frame_denorm = (predictions_frame*(self.target.max()[0]-self.target.min()[0]))+self.target.min()[0]
       
        coef_v, mbe, r2 = get_metrics(predictions_frame_denorm)
            
        return predictions_frame_denorm, coef_v , mbe, r2
            
    def mlp(self,n_iter=100, learning_rate = 0.01,cross_validate=True):
        
        #from sklearn import preprocessing
        #scale=preprocessing.MinMaxScaler()
        #features_scaled = scale.fit_transform(self.features)
        #target_scaled = scale.fit_transform(self.target)
        
        from sknn.mlp import Regressor, Layer
        
        reg = Regressor(layers=[Layer("Rectifier",units=5),  # Hidden Layer1
                                Layer("Rectifier",units=3),  # Hidden Layer2
                                Layer("Linear")],            # Output Layer
                        n_iter = n_iter, learning_rate=learning_rate)
        
        if cross_validate == True:
            predictions_frame = cross_val_regression(reg,
                                                     self.features_norm.as_matrix(),
                                                     self.target_norm.as_matrix(),
                                                     self.series_index_norm,self.rename_col,self.pred_col,self.cv)
            
        else:
            predictions_frame = train_test_regression(reg,
                                                      self.X_train_norm.as_matrix(),
                                                      self.X_test_norm.as_matrix(),
                                                      self.y_train_norm.as_matrix(),
                                                      self.y_test_norm.as_matrix(),
                                                      self.series_index_split_norm,self.rename_col,self.pred_col)
            
        #denormalize the dataframe
        predictions_frame_denorm = (predictions_frame*(self.target.max()[0]-self.target.min()[0]))+self.target.min()[0]
       
        coef_v, mbe, r2 = get_metrics(predictions_frame_denorm)
            
        return predictions_frame_denorm, coef_v , mbe, r2        
```

```python
def cross_val_regression(reg,features,target,series_index,rename_col,pred_col,cv):
    
    from sklearn.model_selection import cross_val_predict
    
    predict = cross_val_predict(reg, features, target,cv=cv)
    predictions = pd.Series(predict.ravel(),index=series_index).rename(rename_col+pred_col)
    return pd.DataFrame(predictions).join(target).dropna()
    
def train_test_regression(reg, X_train, X_test, y_train, y_test,series_index,rename_col,pred_col):
    
    reg.fit(X_train, y_train)
    predict = reg.predict(X_test)
    predictions = pd.Series(predict.ravel(),index=series_index).rename(rename_col+pred_col)
    print y_test
    print type(y_test)
    return pd.DataFrame(predictions).join(pd.Dataframe(y_test,index=series_index)).dropna()
```
